```python
import cv2
import mediapipe as mp
import csv
import os

# initialize
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
hands = mp_hands.Hands(static_image_mode=False,
                       max_num_hands=1,
                       min_detection_confidence=0.5,
                       min_tracking_confidence=0.5)

# file to save data
csv_filename = 'gesture_data.csv'
os.makedirs('data', exist_ok=True)

# create CSV with headers if it doesn't exist
if not os.path.exists(f'data/{csv_filename}'):
    with open(f'data/{csv_filename}', mode='w', newline='') as f:
        writer = csv.writer(f)
        header = []
        for i in range(21):
            header += [f'x{i}', f'y{i}', f'z{i}']
        header += ['label']
        writer.writerow(header)

# start vid
cap = cv2.VideoCapture(0)

print("Press numbers (0-9) to save gestures, with the key representing the gesture label, and 'q' to quit")

while True:
    ret, frame = cap.read()
    if not ret:
        print("done")
        break
        
    
    frame = cv2.flip(frame, 1)
    image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(image_rgb)

    # draw hand landmarks
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

   # show the frame
    cv2.imshow("Gesture Collector", frame)


    # wait for key press
    key = cv2.waitKey(1) & 0xFF

    if key == ord('q'):
        break

    # check if key pressed is a number between 0 and 9
    elif ord('0') <= key <= ord('9'):
        gesture_label = key - ord('0')

        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                data_row = []
                for lm in hand_landmarks.landmark:
                    data_row.extend([lm.x, lm.y, lm.z])
                data_row.append(gesture_label)

                # write row to CSV
                with open(f'data/{csv_filename}', mode='a', newline='') as f:
                    writer = csv.writer(f)
                    writer.writerow(data_row)

                print(f"Saved gesture for label {gesture_label}")
        else:
            print("No hand detected. Try again! :c")

# done
cap.release()
cv2.destroyAllWindows()
```
